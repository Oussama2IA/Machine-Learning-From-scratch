## Importing libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from perceptron import Perceptron
from pocket import Pocket
from adaline import Adaline
```

## Creating a 2d nonlinearly separable dataset

```python
X, y = make_moons(500, noise=0.1, random_state=1)
y = np.where(y == 0, -1, 1)
```

## Visualizing dataset

```python
plt.title('Nonlinearly separable dataset')
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='winter')
plt.show()
```

## Applying machine learning algorithms

### Perceptron

```python
perceptron = Perceptron(max_iters=100)
losses = perceptron.fit(X, y)
hyperplane_params = perceptron.params
```

```python
w = hyperplane_params['w']
b = hyperplane_params['b']
x = np.linspace(X[:, 0].min(), X[:, 0].max())
hyperplane = -(w[0] / w[1]) * x - b / w[1]
plt.title('Perceptron hyperplane')
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='winter')
plt.plot(x, hyperplane, c='red')
plt.show()
```

```python
accuracy_score = perceptron.score(X, y)

print(f'The perceptron model accuracy is {accuracy_score * 100}%')
```

```python
plt.title('Empirical error evolution of Perceptron')
plt.plot(losses)
plt.show()
```

### Pocket

```python
pocket = Pocket(max_iters=100)
losses = pocket.fit(X, y)
hyperplane_params = pocket.params
```

```python
w = hyperplane_params['w']
b = hyperplane_params['b']
x = np.linspace(X[:, 0].min(), X[:, 0].max())
hyperplane = -(w[0] / w[1]) * x - b / w[1]
plt.title('Pocket hyperplane')
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='winter')
plt.plot(x, hyperplane, c='red')
plt.show()
```

```python
accuracy_score = pocket.score(X, y)

print(f'The pocket model accuracy is {accuracy_score * 100}%')
```

```python
plt.title('Empirical error evolution of Pocket')
plt.plot(losses)
plt.show()
```
