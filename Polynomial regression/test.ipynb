### Importing libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

from polynomail_regression import PolynomailRegression
```

### Loading and visualizing data

```python
data = pd.read_csv("data.csv")
X, y = data[['temperature']].values, data['pressure'].values
X = MinMaxScaler().fit_transform(X)
```

```python
plt.xlabel("temperature")
plt.ylabel("pressure")
plt.plot(X, y)
plt.show()
```

*=> The relation between Pressure and Temperature is **not linear** but **curvilinear**.*

### Let's fit the data with polynomial regression model with different polynomial orders (𝑄 = 2, 3, 4)

##### polynomial regression model order 2

```python
pl2 = PolynomailRegression(degree=2)
losses = pl2.fit(X, y)
```

```python
plt.title("The empirical error polynomial regression order 2")
plt.xlabel("Iterations")
plt.ylabel("Loss")
plt.plot(losses)
plt.show()
```

##### polynomial regression model order 3

```python
pl3 = PolynomailRegression(degree=3)
losses = pl3.fit(X, y)
```

```python
plt.title("The empirical error polynomial regression order 3")
plt.xlabel("Iterations")
plt.ylabel("Loss")
plt.plot(losses)
plt.show()
```

##### polynomial regression model order 4

```python
pl4 = PolynomailRegression(degree=4)
losses = pl4.fit(X, y)
```

```python
plt.title("The empirical error polynomial regression order 4")
plt.xlabel("Iterations")
plt.ylabel("Loss")
plt.plot(losses)
plt.show()
```

### 7- Drawing the different polynomial regression model

```python
plt.figure(figsize=(10, 5))
plt.scatter(X, y, color="blue")
plt.plot(X, pl2.predict(X), label="order 2")
plt.plot(X, pl3.predict(X), label="order 3")
plt.plot(X, pl4.predict(X), label="order 4")
plt.title("Polynomial regression with diffrent orders")
plt.xlabel("Temperature")
plt.ylabel("Pressure")
plt.legend()
plt.show()
```
