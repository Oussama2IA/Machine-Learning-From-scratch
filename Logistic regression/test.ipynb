## Importing libraries

```python
import numpy as np
import pandas as pd
from sklearn import datasets
import matplotlib.pyplot as plt
from logistic_regression import LogisticRegression
```

## Accuracy function

```python
def accuracy(y_true, y_pred):
    n = len(y_true)
    S = 0
    for i in range(n):     
        if y_true[i] == y_pred[i]:
            S += 1        
    acc = S / n
    return acc
```

# Creating a 2d dataset

```python
X, y = datasets.make_classification(500, random_state=5)
plt.title('2d dataset')
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='winter')
```

# Logistic regression

```python
lr = LogisticRegression(learning_rate=0.01)
losses = lr.fit(X, y)
hyperplane_params = lr.best_params
```

```python
w = hyperplane_params['w']
b = hyperplane_params['b']
x = np.linspace(X[:, 0].min(), X[:, 0].max())
hyperplane = -(w[0] / w[1]) * x - b / w[1]
plt.title('Logistic regression hyperplane')
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='winter')
plt.plot(x, hyperplane, c='red')
```

```python
y_predicted = lr.predict(X)
accuracy_score = accuracy(y, y_predicted)

print('Model accuracy is:', '{0:.0f}%'.format(accuracy_score * 100))
```
